```python
import pandas as pd
import numpy as np
import requests
```

Postal Codes are at https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_V

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_V'
url_page = requests.get(url)
url_page
```

```python
url_data = pd.read_html(url_page.content)
# url_data
```

```python
len(url_data),type(url_data)
```

```python
wiki_data = url_data[0]
# wiki_data
```

```python
PostalCode = []
Borough = []
Neighbourhood = []
```

```python
for column in wiki_data:
    for row in wiki_data[column]:
        data = row
        parse = data.split('Vancouver',1)
        PostalCode.append(data[0:3])
        Borough.append(parse[0][3:])
        if 'Vancouver' in data:
            Neighbourhood.append(parse[1].replace('/',','))
        else:
            Neighbourhood.append('None')
```

```python
data = {
    'PostalCode':PostalCode,
    'Borough':Borough,
    'Neighbourhood':Neighbourhood
}
```

```python
df = pd.DataFrame(data = data)
df
```

```python
df = df[df.Borough != 'Not assigned']
df.reset_index(inplace=True)
```

```python
del df['index']
```

```python
df
```

```python
df1 = df[df.Neighbourhood != 'None']
df1['Borough'] = df1["Borough"].astype(str) + ' Vancouver'
df1.reset_index(inplace=True)
del df1['index']
df1
```

```python
def get_coor(postalCode):
    coords = None
    while(coords is None):
        g = geocoder.arcgis('{}, Vancouver, British Columbia'.format(postalCode))
        coords = g.latlng
    return coords
get_coor('V5K')
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}'.format(
            client_id, 
            client_secret, 
            version, 
            lat, 
            lng, 
            radius
            )
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Category']
    
    return(nearby_venues)
```

## Each Neighborhood with top 5 most common Venues

```python
values = 5
for hood in Van_grouped['Neighbourhood']:
    print('*'+hood+'*')
    temp = Van_grouped[Van_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue', 'freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(values))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
Van_grouped.shape
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighbourhood'] = Van_grouped['Neighbourhood']

for ind in np.arange(Van_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Van_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## k-means Cluster

```python
from sklearn.cluster import KMeans
```

```python
Van_grouped_clustering = Van_grouped.drop('Neighbourhood', 1)
kmeans = KMeans(n_clusters=10, random_state=0).fit(Van_grouped_clustering)
kmeans.labels_
```

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

Van_merged =df1.iloc[:16,:]

Van_merged = Van_merged.join(neighborhoods_venues_sorted.set_index('Neighbourhood'), on='Neighbourhood')

Van_merged.head()
```
